```python
import pandas as pd
import numpy as np
import csv
```

```python
epa_data = pd.read_csv('epa_data.csv').reset_index(drop=True).drop(columns = ['Unnamed: 0'])
epa_data.tail()
```

```python
list(epa_data)
```

```python
betting_info = pd.read_csv('betting_info copy.csv').reset_index(drop=True).drop(columns = ['Unnamed: 0'])
betting_info = betting_info.dropna(subset = ['spread_line'])
betting_info.tail()
```

```python
list(betting_info)
```

```python
df = betting_info.merge(epa_data, on=['game_id','season','week','home_team','away_team','home_qb_name','away_qb_name'],
 how='outer')
df.fillna('?', inplace = True)

df.tail()
```

```python
list(df)
```

```python
# double check that we're only using regular season games
df = df.loc[df['game_type'] == 'REG']
```

```python
list(df)
```

```python
df = df.drop(columns=[
 'game_type',
 'home_coach',
 'away_coach',
 'div_game',
 'weekday',
 'gametime',
 'referee',
 'stadium',
 'roof',
 'surface',
 'temp',
 'wind',
 'weather',
 'away_team_win',
 'tie',
 'home_ATS_win',
 'road_ATS_win',
 'home_ATS_loss',
 'road_ATS_loss',
 'ATS_push',
 'over',
 'under',
 'push'
 ]
)
```
